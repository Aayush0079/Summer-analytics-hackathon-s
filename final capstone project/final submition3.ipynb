```python
X =df.drop(columns=['price1'],axis=1)
y =df['price1']
```

```python
from pandas.api.types import is_numeric_dtype
```

```python
def get_categorical_indicies(X):
    cats = []
    for col in X.columns:
        if is_numeric_dtype(X[col]):
            pass
        else:
            cats.append(col)
    cat_indicies = []
    for col in cats:
        cat_indicies.append(X.columns.get_loc(col))
    return cat_indicies
categorical_indicies = get_categorical_indicies(X)
```

```python
def convert_cats(X):
    cats = []
    for col in X.columns:
        if is_numeric_dtype(X[col]):
            pass
        else:
            cats.append(col)
    cat_indicies = []
    for col in cats:
        X[col] = X[col].astype('category')
convert_cats(X)
```

```python
from sklearn.model_selection import StratifiedKFold
folds =StratifiedKFold(n_splits=9)#best 9,15,17
```

```python
for train_index, test_index in folds.split(X,y):
   print("TRAIN:", train_index, "TEST:", test_index)
   X_train, X_test,y_train,y_test = X.iloc[list(train_index)], X.iloc[list(test_index)],y.iloc[list(train_index)], y.iloc[list(test_index)]
```

```python
print('Test Survival Rate:',y_test.sum()/y_test.count())
```

```python
train_dataset = cb.Pool(X_train,y_train, 
                        cat_features=categorical_indicies)
```

```python
test_dataset = cb.Pool(X_test,y_test,           
                       cat_features=categorical_indicies)
```

```python
model = cb.CatBoostRegressor(loss_function='RMSE')
```

```python
grid= {'learning_rate': [0.1],
        'depth': [8],
        'iterations': [1000],
        'score_function': ['L2'],
        'border_count' : [256],
        'grow_policy' : ['Depthwise'],
        'reg_lambda': [2],
        'logging_level': ['Silent'],
        'random_strength': [0.8],
        'bagging_temperature': [0.5],
        'sampling_frequency': ['PerTreeLevel'],
        'leaf_estimation_method': ['Gradient'],
        'feature_border_type': ['MaxLogSum'],
        'max_ctr_complexity': [2],
        'fold_len_multiplier': [2]
         }
```

```python

model.grid_search(grid,train_dataset)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
pred = model.predict(X_test)
```

```python
rmse = (np.sqrt(mean_squared_error(y_test, pred)))
```

```python
r2 = r2_score(y_test, pred)
```

```python
print("Testing performance")#226,
print('RMSE: {:.2f}'.format(rmse))
print('R2: {:.2f}'.format(r2))#9 =208 but on kaggle 209
# 7 =206 nd on kaggle 219
```
